#### Slicing String

```python
str1= "Hello Heaven"

# str1[start:end:step]

print(str1[0:5])
print(str1[2:7:2])
```

### FString in Python

```python
name = "Alice"
age = 21                
lang = "Python"
hours = 3

print(f"Hello, {name}. You are {age} years old. You are learning {lang} for {hours} hours every day.")
```

### List in Python

```python
lst_1 = ["apple", "banana", "cherry", "figs"]

print(lst_1[1])
print(lst_1[-1])
lst_1.append("orange") 
print(lst_1)

lst_1.remove("banana")
print(lst_1)

lst_1.append(['grape', 'kiwi'])
print(lst_1)
print(len(lst_1))
lst_1.extend(['grape', 'kiwi'])
print(lst_1)
print(len(lst_1))

lst_1.pop()
```
